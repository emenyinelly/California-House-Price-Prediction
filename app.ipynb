```python
import streamlit as st
import joblib
import numpy as np

st.title("🏠 California House Price Predictor")

# Load model
model = joblib.load('california_house_price_model.pkl')
scaler = joblib.load('scaler.pkl')

col1, col2 = st.columns(2)

with col1:
    medinc = st.number_input("Median Income", value=8.0)
    houseage = st.number_input("House Age", value=41.0)
    ave_rooms = st.number_input("Average Rooms", value=6.0)
    ave_bedrms = st.number_input("Average Bedrooms", value=1.0)

with col2:
    population = st.number_input("Population", value=322)
    ave_occup = st.number_input("Average Occupancy", value=2.5)
    latitude = st.number_input("Latitude", value=37.88)
    longitude = st.number_input("Longitude", value=-122.23)

if st.button("Predict House Price"):
    input_data = np.array([[medinc, houseage, ave_rooms, ave_bedrms,
                          population, ave_occup, latitude, longitude]])
    scaled_input = scaler.transform(input_data)
    prediction = model.predict(scaled_input)[0] * 100000
    st.success(f"Predicted Price: **${prediction:,.2f}**")
```
